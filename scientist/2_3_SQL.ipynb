````python

# Создание и наполнение tab-PC
with open('data/PC.sql', 'w') as file:
    file.write(
        'CREATE TABLE PC (code INTEGER PRIMARY KEY, model INTEGER, speed INTEGER, ram INTEGER, '
        'hd NUMERIC(10,1), cd VARCHAR(50), price NUMERIC(10,4));\n'
    )
    file.write('INSERT INTO PC\n')
    file.write('VALUES\n')

    values = []
    for code, i in enumerate(range(1121, 1225), start=1):
        values.append(
            f'({code}, {i}, {random.randrange(500, 901, 100)},'
            f' {random.randrange(32, 129, 32)}, {random.randrange(5, 21, 5)},'
            f' \'{random.randrange(12, 53, 4)}x\', {random.randrange(350, 1001, 50)})'
        )

    # Чтобы в конце не стояла запятая и код корректно сработал
    file.write(',\n'.join(values) + ';\n')
pc = file
````

#### MariaDB

1. Установка MariaDB:
Для установки MariaDB в Ubuntu 22.04, откройте терминал и выполните следующие команды:
```
sudo apt update
sudo apt install mariadb-server
```

2. Запустите MariaDB:
После успешной установки, запустите MariaDB командой:
```
sudo systemctl start mariadb
```

3. Включите автозапуск MariaDB при загрузке системы:
```
sudo systemctl enable mariadb
```

4. Настройка безопасности MariaDB:
Выполните команду для запуска скрипта настройки безопасности:
```
sudo mysql_secure_installation
```
Следуйте инструкциям на экране для установки пароля администратора, удаления анонимных пользователей, отключения удаленного доступа к базе данных и удаления тестовой базы данных. 

5. Вход в MariaDB:
Чтобы войти в MariaDB и начать работать с базами данных, выполните следующую команду и введите пароль администратора, который вы установили на шаге 4:
```
sudo mysql -u root -p
```

6. Создание новой базы данных:
Чтобы создать новую базу данных, выполните следующую команду в MariaDB:
```
CREATE DATABASE название_базы_данных;
```

7. Создание нового пользователя:
Чтобы создать нового пользователя и предоставить ему права на доступ и управление базой данных, выполните следующие команды:
```
CREATE USER 'имя_пользователя'@'localhost' IDENTIFIED BY 'пароль';
GRANT ALL PRIVILEGES ON название_базы_данных.* TO 'имя_пользователя'@'localhost';
FLUSH PRIVILEGES;
```

Это основные шаги по установке и настройке MariaDB в Ubuntu 22.04. Следуйте этим инструкциям и вы сможете успешно установить и настроить MariaDB на своем сервере.

#### psycopg2


Библиотека psycopg2

Psycopg2 - это библиотека Python, которая предоставляет интерфейс программирования приложений (API) для доступа к PostgreSQL. Она поддерживает большинство современных версий PostgreSQL.

Psycopg2 был разработан для обеспечения высокой производительности и совместимости с различными версиями PostgreSQL. Он также предоставляет поддержку для различных типов данных, таких как строки, числа, даты и время, а также массивы.

Одним из преимуществ psycopg2 является его простота использования. Он имеет простой и понятный API, который позволяет разработчикам быстро начать работу с PostgreSQL. Кроме того, psycopg2 поддерживает множество других функций, таких как транзакции, курсоры, параметры команд и многое другое.

В целом, psycopg2 является отличным выбором для разработчиков Python, которые хотят работать с PostgreSQL. Он обеспечивает высокую производительность, простоту использования и широкий спектр функций, которые делают его идеальным для различных приложений.

5. Создаем контейнер для pgAdmin

    (через командную строку) вбиваем команду: docker inspect myPostgreSQL (название указано выше)

    копируем IP
