Задача 1. Создайте программу, запрашивающую у пользователя два целых числа $a$ и $b$, ввод которых осуществляется через символ пробела, т.е. числа $a$ и $b$ вводятся в одной строке через пробел. Выведите на экран результаты следующих математических операций (предполагается, что соответствующие операции определены):
- сумма `a` и `b`;
- разность `a` и `b`;
- произведение `a` и `b`;
- частное от деления `a` на `b`;
- остаток от деления `a` на `b`;
- десятичный логарифм числа `a;
- результат возведения числа `a` в степень `b`;
- квадратный корень из числа `a`;
- результат умножения числа $e$ на `b`.

Примечание: рекомендуется
- для ввода чисел использовать функцию [`input()`](https://pythonru.com/osnovy/poluchenie-polzovatelskogo-vvoda-v-python-s-input);

Напомним, что подключить библиотеку можно следующим способом
```python

```

````python
import math
from collections import Counter
import re
````

````python
input_str = input("Введите два целых числа a и b (через пробел): ")
a, b = map(int, input_str.split())

# Вычисляем и выводим результаты операций
print(f"Сумма a и b: {a + b}")
print(f"Разность a и b: {a - b}")
print(f"Произведение a и b: {a * b}")

# Проверяем, чтобы b не было равным нулю для операций с делением
if b != 0:
    print(f"Частное a на b: {a / b}")
    print(f"Остаток от деления a на b: {a % b}")
else:
    print("Деление на ноль невозможно, операция пропущена.")

# Проверяем, чтобы a было больше нуля для логарифма и корня
if a > 0:
    print(f"Десятичный логарифм числа a: {math.log10(a)}")
    print(f"Квадратный корень из числа a: {math.sqrt(a)}")
else:
    print("Логарифм и квадратный корень определены только для положительных чисел a.")

# Выводим результат возведения a в степень b
print(f"a в степени b: {a ** b}")

# Умножение числа e на b
print(f"Число e, умноженное на b: {math.e * b}")
````

Задача 2. Имеются две пиццы диаметрами 35 см. и 40 см., соответственно. Насколько процентов отличаются площади двух пицц?

Напомним, что площадь круга вычисляется по следующей формуле
$$S = \pi r^2,$$
где $r$ -- радиус круга.

````python
diam_1 = 35  
diam_2 = 40  

rad_1 = diam_1 / 2
rad_2 = diam_2 / 2

s_1 = math.pi * rad_1 ** 2
s_2 = math.pi * rad_2 ** 2

percent = ((s_2 - s_1) / s_2) * 100

print(f"Различие % в площадях пицц: {percent:.2f}%")
````

Примечание: методы для работы со списками можно посмотреть [здесь](https://pythonworld.ru/tipy-dannyx-v-python/spiski-list-funkcii-i-metody-spiskov.html).

````python
# список чисел
numbers = list(range(1, 20, 3))

# длинa списка
print("длина списка:", len(numbers))

# все элементы списка, начиная со 2
print('начиная со 2-го', numbers[1:])

# все четные 
even_numbers = [num for num in numbers if num % 2 == 0]
print("четные элементы:", even_numbers)

# элементы списка, делящиеся на 5 без остатка
divisible_by_5 = [num for num in numbers if num % 5 == 0]
if divisible_by_5:
    print("делящиеся на 5:", divisible_by_5)
else:
    print("Элементов, делящихся на 5, нет.")

# Отсортируйте в порядке убывания
numbers.sort(reverse=True)
print("отсортированный список в порядке убывания:", numbers)

# квадраты элементов с использованием map и pow
squares = list(map(lambda x: pow(x, 2), numbers))
print("квадраты элементов списка:", squares)

# элементы списка в обратном порядке
print("Элементы списка в обратном порядке:", numbers[::-1])

# выборочное среднее
sum_elements = sum(numbers)
average = sum_elements / len(numbers) if numbers else 0
print("Выборочное среднее элементов списка:", average)
````

````python
data = {
    'понедельник': 12.0,
    'вторник': 12.0,
    'среда': 17.0,
    'четверг': 17.0,
    'пятница': 14.3,
    'суббота': 11.9,
    'воскресенье': 12.0
}
data
````

````python
# температура Фаренгейту
for i in data:
    print(i, 'Цельсий', data[i], 'Фаренгейт:', (data[i] * 9 / 5) + 32)
````

````python
# Дни с максимальной и минимальной температурой
max_temp = max(data.values())
min_temp = min(data.values())

max_days = [day for day, temp in data.items() if temp == max_temp]
min_days = [day for day, temp in data.items() if temp == min_temp]

print(f"Максимальная температура: {max_temp} °C, наблюдалась в дни: {max_days}")
print(f"Минимальная температура: {min_temp} °C, наблюдалась в дни: {min_days}")
````

````python
# Среднее значение температуры за неделю
temp = sum(data.values()) / len(data)
print(f"Среднее значение температуры за неделю: {temp:.2f} °C")
````

````python
# Модальное значение температуры
temp_counts = Counter(data.values())
mode_temp = temp_counts.most_common(1)[0][0] 
mode_days = [day for day, temp in data.items() if temp == mode_temp]

print(f"Модальное значение температуры: {mode_temp} °C, наблюдалась в дни: {mode_days}")
````

````python
# Прохладные и теплые дни
cool_days = [day for day, temp in data.items() if temp < 14]
warm_days = [day for day, temp in data.items() if temp >= 14]

print(f"Прохладные дни: {cool_days}")
print(f"Теплые дни: {warm_days}")
````

Задача 6. Имеются следующие списки
```python
>>> disc = ['Алгебра', 'Геометрия', 'Топология']
>>> student = ['Сидоров', 'Иванов', 'Петров', 'Иванова']
>>> grades_student = [[2, 4, 4], [3, 4, 4], [3, 3, 3], [5, 5, 5]]
```
Здесь

|Список|Комментарий|
|--:|:--|
|`disc`|Перечень экзаменов по дисциплинам|
|`student`|Фамилии студентов|
|`grades_student`|Оценки студентов по экзаменам|

Необходимо из имеющихся списков:
1. С помощью методов `dict()` и `zip()` сформируйте словарь вида
```python
{'Иванов': [3, 4, 4],
 'Иванова': [5, 5, 5],
 'Петров': [3, 3, 3],
 'Сидоров': [2, 4, 4]}
```
2. С помощью цикла `for` и метода `enumerate()` cформируйте словарь вида
```python
{'Иванов': {'Алгебра': 3, 'Геометрия': 4, 'Топология': 4},
 'Иванова': {'Алгебра': 5, 'Геометрия': 5, 'Топология': 5},
 'Петров': {'Алгебра': 3, 'Геометрия': 3, 'Топология': 3},
 'Сидоров': {'Алгебра': 2, 'Геометрия': 4, 'Топология': 4}}
```
3. В словаре, полученном во 2-м пункте, найдите и распечатайте фамилии <<круглых отличников>> и  студентов, несдавших сессию.

````python
disc = ['Алгебра', 'Геометрия', 'Топология']
student = ['Сидоров', 'Иванов', 'Петров', 'Иванова']
grades_student = [[2, 4, 4], [3, 4, 4], [3, 3, 3], [5, 5, 5]]

# Формируем словарь dict() и zip()
grades_dict = dict(zip(student, grades_student))
print("1. Словарь оценок студентов:")
print(grades_dict)
````

````python
# Формируем словарь с дисциплинами и оценками
grades_detailed_dict = {}
for i, name in enumerate(student):
    grades_detailed_dict[name] = {disc[j]: grades_student[i][j] for j in range(len(disc))}

print("\n2. Словарь оценок студентов по дисциплинам:")
print(grades_detailed_dict)
````

````python
# Ищем круговых отличников и несдавших сессию
good_students = []
bad_students = []

for i, grades in grades_detailed_dict.items():
    if all(grade == 5 for grade in grades.values()):
        good_students.append(i)
    if any(grade < 4 for grade in grades.values()):
        bad_students.append(i)

print("\n3. Круглые отличники:")
print(good_students)
print("Несдавшие сессию:")
print(bad_students)
````

Задача 7. Целое положительное число из отрезка $[3;50]$ вводится с клавиатуры. Найдите сумму всех положительных четных чисел, строго меньших введенного.

Примечание: здесь можно предложить несколько решений, например, только через списки или с использованием циклов. В целях ознакомления и овладения навыками работы с циклами рекомендуется использовать цикл с предусловием `while`.

````python
num = int(input("Введите целое положительное число от 3 до 50: "))

# Проверка, что введенное число находится в нужном диапазоне
if 3 <= num <= 50:
    sum_even = 0
    current_even = 2  # Начинаем с первого четного числа

    # Используем цикл while для суммирования четных чисел
    while current_even < num:
        sum_even += current_even
        current_even += 2  # Переход к следующему четному числу

    # Выводим результат
    print("Сумма всех положительных четных чисел, строго меньших", num, ":", sum_even)
else:
    print("Введите число в диапазоне от 3 до 50!")
````

````python
text = '''
Взвейтесь кострами, синие ночи!

Мы пионеры - дети рабочих

Близится время светлых годов

Клич пионера: "Всегда будь готов!"

Радостным шагом с песней веселой

Мы выступаем за комсомолом

Близится время светлых годов

Клич пионера: "Всегда будь готов!"

Мы поднимаем красное знамя

Дети рабочих, смело за нами!

Близится время светлых годов

Клич пионера: "Всегда будь готов!"

Взвейтесь кострами, синие ночи!

Мы пионеры - дети рабочих

Близится время светлых годов

Клич пионера: "Всегда будь готов!"

Клич пионера: "Всегда будь готов!"
'''
````

````python
# к нижнему регистру
text_lower = text.lower()

# строк
lines_count = text_lower.strip().count('\n') + 1  # +1 для последней строки

# предложений
sentences_count = len(re.findall(r'[.!?]', text_lower))

# Уникальные слова
# Удаляем знаки препинания, делим текст на слова и создаем множество уникальных слов
words = re.findall(r'\b\w+\b', text_lower)
unique_words = set(words)

# Частотная таблица для уникальных слов
word_frequency = Counter(words)

# Уникальные буквы
unique_letters = set(c for c in text_lower if c.isalpha())
````

````python
print(f'Количество строк: {lines_count}')
print(f'Количество предложений: {sentences_count}')
print(f'Количество уникальных слов: {len(unique_words)}')
print('Частотная таблица для каждого уникального слова:')
for word, count in word_frequency.items():
    print(f'{word}: {count}')
print(f'Количество уникальных букв: {len(unique_letters)}')
````
