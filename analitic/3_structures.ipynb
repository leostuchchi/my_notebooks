**Задание 1.** Напишите функцию для нахождения площади круга.

Напомним, что площадь круга вычисляется по следующей формуле
$$S = \pi r^2,$$
где $r$ -- радиус круга.

Формат выполнения задания:

```python
#определение функции
def f.....

#вызов функции и результат ее работы
f(1)
Площадь круга радиуса 1 равна 3.14 кв.ед.
```



````python
import math as m
````

````python
def f(r):
    s = round((m.pi * r **2), 2)
    #return 'Площадь круга радиуса ', round(r, 2), 'равна', s, ' кв.ед.'

    return f'Площадь круга радиуса {r} равна {s} кв.ед.'
````

````python
f(1)
````

Задание 2. Напишите функцию для нахождения суммы `n` элементов списка.  Если `n` не задано, то функция должна возвращать сумму всех элементов списка. Список значений и `n` задаются пользователем произвольно (либо через клавиатуру, либо через переменные).

Формат выполнения задания:

```python
#определение функции
def s.....

#вызов функции и результат ее работы
list_example = range(-10, 10)
n = 2
s(list_example, n)
Сумма элементов -10, -9 равна -19.

s(list_example)
Сумма всех элементов списка равна -10.
```


````python
def s(lst, n=None):
    if n:
        result = sum(lst[:n])
        print(f"Сумма {n} элементов списка равна {result}.")
    else:
        result = sum(lst)
        print(f"Сумма всех элементов списка равна {result}.")
````

````python
list_example = range(-10, 10)
n = 2
s(list_example, n)
````

````python
list_example = range(-10, 10)
n = 2
s(list_example)
````

Задание 3. Пусть общий доход складывается только из следующих источников: заработной платы, доходов по вкладам, дивидентов, пособия. Напишите функцию, рассчитывающую для Вас общий доход. Предполагается, что некоторых источников дохода у Вас может не быть. Попробуйте задать параметры в виде 1) кортежа; 2) словаря.


Формат выполнения задания:

```python
#определение функций
def income.....

#вызов функциb и результат ее работы
income(100)
Общий доход равен 100.

income(100, 10)
Общий доход равен 1010.
```

````python

def income(*args):
    total_income = sum(args)
    print(f"Общий доход равен {total_income}.")
````

````python
income(100)
````

````python
income(100, 10, 35)
````

````python

def income_dict(**kwargs):
    total_income = sum(kwargs.values())
    print(f"Общий доход равен {total_income}.")
income_dict(salary=100, deposits=200, dividends=300)
````

````python
income_dict(salary=100, deposits=10, dividends=50, allowance=20)
````

````python
def income(*args, **kwargs):
    total_income = sum(args) + sum(kwargs.values())
    print(f"Общий доход равен {total_income}.")
````

````python
income(100, 200, 300, salary=100, deposits=200, dividends=300)
````
